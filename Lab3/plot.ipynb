```python
import matplotlib.pyplot as plt
import json
```

```python

with open('top_tokens.json', 'r') as f:
    data = json.load(f)
```

```python
x = list(data.keys())
y = list(data.values())
```

```python
k=90
plt.bar(x[k:100], y[k:100], color='maroon')  
plt.xlabel('x-axis')
plt.ylabel('Frequency')
plt.title('Frequency Bar Plot from JSON file')
plt.show()
```
