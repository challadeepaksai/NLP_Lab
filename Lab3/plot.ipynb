```python
import matplotlib.pyplot as plt
import json
```

```python

with open('top_tokens.json', 'r') as f:
    data = json.load(f)
```

```python
x = list(data.keys())
y = list(data.values())
```

```python
plt.plot(x[:10], y[:10], marker='o', color='maroon')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot from JSON file')
plt.show()
```
